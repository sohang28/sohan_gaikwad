```python
# remove space between column labels and add _(underscore)
col=list(f.columns)
c1=[]
for i in col:
    i=i.replace(' ','_')
    c1.append(i)
c1
f.columns=c1
f.head(2)
```

```python
# remove ID column
f.drop(columns='ID',inplace=True)
f.head(5)
```

```python
# Display data of FC Barcelona
fcb=f.query('Club=="FC Barcelona"').sort_values('Overall',ascending=False)
fcb
```

```python
# fetch player whose age is >30 & count how many are there
a=f.query('Age>30')
a
```

```python
a.shape[0]
```

```python
# pie plot for preferred foot and give your analysis
PF=f.Preferred_Foot.value_counts()
PF
```

```python
PF.plot(kind='pie',autopct='%.2f%%')
```

Analysis:

1- From the pie chart we can say that 76.87% i.e 13996 players preferred right foot which is in blue color.

2- and 23.13% i.e 4211 players preferred left foot to play that we can see in orange color.
